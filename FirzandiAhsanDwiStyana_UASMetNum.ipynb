```julia
#Bagian (a) - Pasangan untuk titik-titik (2, 2) dan (6, 2).

using Plots

x1 = [2,6]
y1 = [2,2]
plot(x1,y1,label="Bagian (a): (2, 2) dan (6, 2)", xlim=(0, 8), ylim=(0, 15), color = "fuchsia")

xlabel!("X")
ylabel!("Y")
```

```julia
#Bagian (b) - Pasangan untuk titik-titik (2, 2) dan (6, 4).

using Plots

x2 = [2,6]
y2 = [2,4]
plot!(x2,y2,label="Bagian (b): (2, 2) dan (6, 4)", color = "sandybrown")

xlabel!("X")
ylabel!("Y")
```
